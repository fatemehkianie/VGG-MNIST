```python
from tensorflow.keras.datasets.mnist import load_data
```

```python
(trainX, trainy), (testX, testy) = load_data()
trainX = trainX[:1000]
trainy = trainy[:1000]
testX = testX[:100]
testy = testy[:100]
```

```python
import tensorflow as tf
trainX = tf.map_fn(lambda i: tf.stack([i]*3, axis=-1), trainX).numpy()
testX = tf.map_fn(lambda i: tf.stack([i]*3, axis=-1), testX).numpy()
#Resizing
trainX = tf.image.resize(trainX, [224,224]).numpy()
testX = tf.image.resize(testX, [224,224]).numpy()
```

```python
trainy =tf.keras.utils.to_categorical(trainy, num_classes=10)
testy =tf.keras.utils.to_categorical(testy, num_classes=10)
```

```python
trainX= trainX/255.0
testX= testX/255.0
```

```python
INDEX = int(0.8*len(trainX))

trainX = trainX[:INDEX]
trainy = trainy[:INDEX]

val_x = testX[INDEX:]
val_y = testy[INDEX:]
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import MaxPool2D
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dense
```

Step1: Define the model

```python
VGG16 = Sequential()
VGG16.add(Conv2D(64, (3,3), padding='same', activation='relu', input_shape=(224,224,3)))
VGG16.add(Conv2D(64, (3,3), padding='same', activation='relu'))
VGG16.add(MaxPool2D(pool_size=(2,2), strides=(2,2)))
VGG16.add(Conv2D(128, (3,3), padding='same', activation='relu'))
VGG16.add(Conv2D(128, (3,3), padding='same', activation='relu'))
VGG16.add(MaxPool2D(pool_size=(2,2), strides=(2,2)))
VGG16.add(Conv2D(256, (3,3), padding='same', activation='relu'))
VGG16.add(Conv2D(256, (3,3), padding='same', activation='relu'))
VGG16.add(Conv2D(256, (3,3), padding='same', activation='relu'))
VGG16.add(MaxPool2D(pool_size=(2,2), strides=(2,2)))
VGG16.add(Conv2D(512, (3,3), padding='same', activation='relu'))
VGG16.add(Conv2D(512, (3,3), padding='same', activation='relu'))
VGG16.add(Conv2D(512, (3,3), padding='same', activation='relu'))
VGG16.add(MaxPool2D(pool_size=(2,2), strides=(2,2)))
VGG16.add(Conv2D(512, (3,3), padding='same', activation='relu'))
VGG16.add(Conv2D(512, (3,3), padding='same', activation='relu'))
VGG16.add(Conv2D(512, (3,3), padding='same', activation='relu'))
VGG16.add(MaxPool2D(pool_size=(2,2), strides=(2,2)))
VGG16.add(Flatten())
VGG16.add(Dense(4096, activation='relu'))
VGG16.add(Dense(4096, activation='relu'))
VGG16.add(Dense(10, activation='softmax'))
```

Step2: Compile the model

```python
VGG16.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=0.01, momentum=0.9), 
              loss ='categorical_crossentropy',
              metrics=['accuracy', tf.keras.metrics.TopKCategoricalAccuracy(5)])

reduce_lr = tf.keras.callbacks.ReduceLROnPlateau(monitor='loss', factor=0.1, patience=1, min_lr=0.00001)
```

Step3: Fit the model

```python
history = VGG16.fit(trainX, trainy, batch_size=128, validation_data =(val_x, val_y), epochs=10, callbacks=[reduce_lr])
```
